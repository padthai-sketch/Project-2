```python
dfs = []
for x in range(52):
    time.sleep(.01)
    offset = int(x*200)
    url=f'https://api.bridgedataoutput.com/api/v2/actris_ref/listings?access_token=b36573f7738db93c5e2a5b9171c3626a&offset={offset}&limit=200'
    try:
        df1 = pd.read_json(url)
        dfs.append(df1[:])
    except:
        break

df = pd.concat(dfs, axis=0)
display(df.head())
len(df)
```

```python
df1 = pd.json_normalize(df['bundle'])
df1.head()
```

```python
drop = ['GreenSustainability', 'OtherEquipment', 'url', 'BuyerOfficeKey', 'CoBuyerAgentFullName', 'AvailabilityDate', 'ListAgentKey', 'TaxLegalDescription', 'StatusChangeTimestamp', 'RoadSurfaceType', 'PriceChangeTimestamp', 'ListingContractDate', 'CurrentUse', 'OriginatingSystemKey', 'PhotosChangeTimestamp', 'OriginatingSystemName', 'MlsStatus', 'ListingKey', 'StateOrProvince', 'ListAgentKeyNumeric', 'SyndicateTo', 'ListOfficeKeyNumeric', 'StreetDirSuffix', 'Directions', 'BridgeModificationTimestamp', 'ParcelNumber', 'OperatingExpenseIncludes', 'SubdivisionName', 'OffMarketDate', 'CloseDate', 'HorseAmenities', 'ModificationTimestamp', 'BuyerAgentKey', 'OriginatingSystemID', 'UnitNumber', 'StreetDirPrefix', 'LivingAreaUnits', 'ContingentDate', 'AssociationName', 'BuildingName', 'CoBuyerAgentKey', 'CoListOfficeName', 'WaterBodyName', 'AboveGradeFinishedArea', 'ListingKeyNumeric', 'MajorChangeType', 'LeasableArea', 'CoListOfficeKey', 'Zoning', 'CoBuyerOfficeName', 'TaxBlock', 'DualVariableCompensationYN', 'WaterBodyName', 'GrossScheduledIncome', 'BelowGradeFinishedAreaSource', 'CoBuyerOfficeKey', 'LeaseTerm', 'AboveGradeFinishedAreaUnits', 'VirtualTourURLUnbranded', 'ListOfficeKey', 'StandardStatus', 'CultivatedArea', 'ParkingTotal', 'CoListOfficeKeyNumeric', 'AdditionalParcelsDescription', 'OperatingExpense', 'InternetAddressDisplayYN', 'Contingency', 'ListingService', 'Topography', 'PropertyAttachedYN', 'TaxLot', 'NetOperatingIncome', 'OtherStructures', 'VirtualTourURLZillow', 'CoListAgentFullName', 'CoListAgentKey', 'BelowGradeFinishedAreaUnits', 'EntryLocation', 'TaxAssessedValue', 'CoListAgentKeyNumeric', 'ExistingLeaseType', 'PostalCodePlus4', 'TotalActualRent', 'StreetName', 'OwnerPays', 'AnchorsCoTenants', 'ListingAgreement', 'ListingId', 'DocumentsChangeTimestamp', 'WindowFeatures', 'SpecialListingConditions', 'City', 'InternetEntireListingDisplayYN', 'InternetAutomatedValuationDisplayYN', 'BuildingFeatures', 'OwnershipType', 'DocumentsAvailable', 'NumberOfUnitsTotal', 'TenantPays', 'InternetConsumerCommentYN', 'DocumentsCount', 'StoriesTotal', 'NumberOfBuildings']
df2 = df1.drop(drop, axis=1)
df2.head()
```

```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# Download/Update the VADER Lexicon
nltk.download('vader_lexicon')
nltk.download('stopwords')

# Initialize the VADER sentiment analyzer
analyzer = SentimentIntensityAnalyzer()

# Create the Facebook Libra sentiment scores DataFrame
sentiments = []

for text in df2["PublicRemarks"]:
    try:
        sentiment = analyzer.polarity_scores(text)
        compound = sentiment["compound"]
        pos = sentiment["pos"]
        neu = sentiment["neu"]
        neg = sentiment["neg"]
        
        sentiments.append({
            "compound": compound,
            "positive": pos,
            "negative": neg,
            "neutral": neu
            
        })    
    except AttributeError:
        pass
    
# Create DataFrame
df3 = pd.DataFrame(sentiments)

# Reorder DataFrame columns
cols = ["compound", "positive", "negative", "neutral"]
df3 = df3[cols]

df3.head()
df3 = pd.concat([df2, df3], axis=1)
df3.drop(['PublicRemarks'], axis=1, inplace=True)
df3.head()
```

```python
hot_encode = ['BuyerAgentFullName', 'ListOfficeName', 'MLSAreaMajor', 'PropertyType', 'CountyOrParish', 'MiddleOrJuniorSchool', 'ElementarySchool', 'ListAgentFullName', 'PropertySubType', 'HighSchool', 'HighSchoolDistrict', 'StreetSuffix', 'AssociationFeeFrequency', 'ElementarySchoolDistrict', 'OccupantType', 'DirectionFaces', 'Furnished', 'BuyerOfficeName', 'MiddleOrJuniorSchoolDistrict']
df4 = pd.get_dummies(df3, columns=hot_encode)
df4.head()
```

```python
tru_false = ['AdditionalParcelsYN', 'GarageYN', 'AssociationYN', 'AttachedGarageYN', 'PoolPrivateYN', 'IDXParticipationYN', 'WaterfrontYN', 'HorseYN', 'ViewYN', 'NewConstructionYN']
for col in tru_false:
    df4[col] = np.where(df4[col] == True, 1, 0)
df4.head()
```

```python
col = 'LotSizeDimensions'
df4[col] = df4[col].apply(lambda x: re.sub("[^0-9]", "", str(x)))
df4[df4[col] == ''] = 0
df4[col] = df4[col].astype(float)
print(df4[col].values)

col = 'StreetNumber'
df4[col] = df4[col].apply(lambda x: re.sub("[^0-9]", "", str(x)))
df4[df4[col] == ''] = 0
df4[col] = df4[col].astype(float)
print(df4[col].values)
```

```python
to_int = ['PostalCode']
for col in to_int:
    df4[col] = df4[col].astype(int)

df4.head()
```

```python
df5 = df4[:]
expand = ['WaterfrontFeatures', 'Flooring', 'Cooling', 'ExteriorFeatures', 'SecurityFeatures', 'Appliances', 'AssociationFeeIncludes', 'Roof', 'PropertyCondition', 'ListingTerms', 'FoundationDetails', 'View', 'Levels', 'InteriorFeatures', 'LockBoxType', 'Fencing', 'WaterSource', 'Disclosures', 'PetsAllowed', 'SpaFeatures', 'ParkingFeatures', 'PoolFeatures', 'Sewer', 'Heating', 'PatioAndPorchFeatures', 'Utilities', 'FireplaceFeatures', 'CommunityFeatures', 'GreenEnergyEfficient', 'AccessibilityFeatures', 'LotFeatures', 'ConstructionMaterials']

for col in expand:
    tags = df5[col].apply(pd.Series)
    tags = tags.rename(columns = lambda x : col + '_' + str(x))
    tags.fillna(0, inplace=True)
    tags[tags != 0] = 1
    df5 = pd.concat([df5, tags], axis=1)
    
df5.drop(expand, inplace=True, axis=1)
```

```python
df5.head()
```

```python
media_series = df5.pop('Media')
address_series = df5.pop('UnparsedAddress')

df5.fillna(0, inplace=True)
df5 = df5.astype(int)

for col in df5.columns:
    if len(df5[df5[col] > 0]) == 0:
        df5.drop([col], inplace=True, axis=1)
```
